Get the movie names and years into lists and only leave the movies wihtout repetition.

```python
import pandas as pd
df = pd.read_csv('movie.csv')
df.head()
```

```python
dfnew = df.drop_duplicates(subset = ['3', '5'],keep = 'last').reset_index(drop = True)
dfnew
```

```python
dfnew = dfnew.rename(columns={"3": "movie_name", "5": "year"})
dfnew = dfnew.drop(columns=['Unnamed: 0','0', '1','2','4'])
dfnew = dfnew.astype({"year": str})
dfnew
```

```python
movie_list = dfnew.to_dict('records')
movie_list
```

Try to get one movie url

```python
url = "https://www.imdb.com/find?q=The+New+World+2004&ref_=nv_sr_sm"
await page.goto(url)
```

```python
from bs4 import BeautifulSoup
html = await page.content()
soup_doc = BeautifulSoup(html, "html.parser")
link = "https://www.imdb.com"+soup_doc.find(class_ = 'ipc-page-section ipc-page-section--base sc-17bafbdb-0 cuASnq').find_all(class_ = 'ipc-metadata-list-summary-item ipc-metadata-list-summary-item--click find-result-item find-title-result')[0].a.get('href')
link
```

loop through everything to get all the urls

```python
for movie in movie_list:
    await page.goto(movie['search_url'])
    from bs4 import BeautifulSoup
    html = await page.content()
    soup_doc = BeautifulSoup(html, "html.parser")

    try:
        link = "https://www.imdb.com"+soup_doc.find(class_ = 'ipc-page-section ipc-page-section--base sc-17bafbdb-0 cuASnq').find_all(class_ = 'ipc-metadata-list-summary-item ipc-metadata-list-summary-item--click find-result-item find-title-result')[0].a.get('href')
        movie['movie_url'] = link
        print(link,"found")
    except:
        movie['movie_url'] = 'ERROR'
        print("error appending", movie['search_url'])

movie_list
```

use playwright to access all the urls, and get the production location

```python
for movie in movie_list:
    
    try:
        await page.goto(movie['movie_url'])
        from bs4 import BeautifulSoup
        html = await page.content()
        soup_doc = BeautifulSoup(html, "html.parser")

        location = soup_doc.select('[data-testid="Details"]')[0].select('[data-testid="title-details-filminglocations"]')[0].li.text
        movie['filming_location'] = location
        print(location,"found")
    except:
        movie['filming_location'] = 'ERROR'
        print("error appending", movie['movie_name'])

movie_list
```
