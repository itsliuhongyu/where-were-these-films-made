```python
df = df[df['filming_location'] != 'ERROR']
df
```

```python
def make_country(cell):
    if "," in cell:
        return cell.split(",")[-1]
    else:
        return cell
```

```python
df['filming_country'] = df['filming_location'].apply(make_country)
```

```python
df
```

```python
df['filming_country'] = df['filming_country'].str.strip()
replacers = {"USA":"United States of America","UK":"United Kingdom","Norge":"Norway","Czech Republic":"Czechia","California USA":"United States of America","Bosnia-Herzegovina":"Bosnia and Herzegovina"}
df['filming_country'] = df['filming_country'].replace(replacers)
df
```
